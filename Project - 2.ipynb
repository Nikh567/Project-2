```python
category_df=df[['MaritalStatus','BusinessTravel','JobRole','Department','EducationField','Gender']]
category_df
```

```python
from sklearn.preprocessing import OneHotEncoder as OHE
onehot=OHE()
category_df=onehot.fit_transform(category_df).toarray()
```

```python
number_df=df[['Age','DailyRate','DistanceFromHome','Education','EnvironmentSatisfaction','HourlyRate','JobInvolvement'
              ,'JobLevel','JobSatisfaction','MonthlyIncome','MonthlyRate','NumCompaniesWorked','OverTime','PercentSalaryHike','PerformanceRating',	
              'RelationshipSatisfaction','StockOptionLevel','TotalWorkingYears'	,'TrainingTimesLastYear'	,'WorkLifeBalance',
              'YearsAtCompany','YearsInCurrentRole','YearsSinceLastPromotion','YearsWithCurrManager']]
number_df.head()
```

```python
df.info()
```

```python
from sklearn.model_selection import train_test_split as tt
category_df=pd.DataFrame(category_df)
x=pd.concat([number_df,category_df],axis=1)
x
```

```python
from sklearn.preprocessing import MinMaxScaler as MMS
first_pre=MMS()
x=first_pre.fit_transform(x)
x
```

```python
pd.DataFrame(x)
```

```python
y=df['Attrition']
y
```

```python
x_train,x_test,y_train,y_test=tt(x,y,test_size=0.25,random_state=23)
```

```python
x_train.shape
```

```python
x_train.ndim
```

```python
y_train.ndim
```

```python
x.dtype
```

```python
from sklearn.linear_model import LogisticRegression as LR
from sklearn.metrics import confusion_matrix as CM,classification_report as CR
first_mod=LR()
first_mod.fit(x_train,y_train)
y_pred=first_mod.predict(x_test)
print(CM(y_test,y_pred))
print(CR(y_test,y_pred))
```

```python
print('The Accuracy between Actual and Modelled Data is ',(305+26)/(305+26+34+3)*100.0,'%')
```

```python
from sklearn.ensemble import RandomForestClassifier as RFC
sec_mod=RFC()
sec_mod.fit(x_train,y_train)
y_pred2=sec_mod.predict(x_test)
print(CM(y_test,y_pred2))
print(CR(y_test,y_pred2))
```

```python
print('The Accuracy between the Actual and Predicted data is ',(301+9)/(307+10+51)*100,'%')
```

```python
from xgboost import XGBClassifier as xgbc
third_mod=xgbc()
third_mod.fit(x_train,y_train)
y_pred3=third_mod.predict(x_test)
print(CM(y_test,y_pred3))
```

```python
print('The Accuracy between Actual and Predicted data is : ',(301+17)/(301+50+17)*100,'%')
```
